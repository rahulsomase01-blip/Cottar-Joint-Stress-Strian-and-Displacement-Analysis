```python
# Google Colab / Python version of your MATLAB code

# Enter the axial load P (in N): 50000
# Enter the yield strength Sy (in N/mm^2): 400
# Enter Factor of Safety for rod/spigot/socket: 6
# Enter Factor of Safety for cotter: 4
# use as per your design
import math

# User Inputs
P = float(input("Enter the axial load P (in N): "))
Sy = float(input("Enter the yield strength Sy (in N/mm^2): "))
fos_rod = float(input("Enter Factor of Safety for rod/spigot/socket: "))
fos_cotter = float(input("Enter Factor of Safety for cotter: "))

# Permissible Stresses
sigma_t_rod = Sy / fos_rod
sigma_c_perm = 2 * Sy / fos_rod
tau_rod = 0.5 * Sy / fos_rod

tau_cotter = 0.5 * Sy / fos_cotter
sigma_t_cotter = Sy / fos_cotter

print("\n--- Permissible Stresses in Rod, Spigot, Socket ---")
print(f"σt_rod = {sigma_t_rod:.4f} N/mm²")
print(f"σc = {sigma_c_perm:.4f} N/mm²")
print(f"τ_rod = {tau_rod:.4f} N/mm²")

print("\n--- Permissible Stresses in Cotter ---")
print(f"τ_cotter = {tau_cotter:.4f} N/mm²")
print(f"σt_cotter = {sigma_t_cotter:.4f} N/mm²")

# Step 1: Rod Diameter
d = math.sqrt((4 * P) / (math.pi * sigma_t_rod))
print(f"\nRod Diameter (d): {d:.4f} mm")

# Step 2: Cotter Thickness
t = 0.31 * d
print(f"Cotter Thickness (t): {t:.4f} mm")

# Step 3: Solve for d2 (Quadratic equation)
a = math.pi / 4
b = -t
c = -P / sigma_t_rod

discriminant = b**2 - 4 * a * c

if discriminant >= 0:
    root1 = (-b + math.sqrt(discriminant)) / (2 * a)
    root2 = (-b - math.sqrt(discriminant)) / (2 * a)
    d2_values = [r for r in [root1, root2] if r > 0]

    if not d2_values:
        print("Error: No positive solution for d2.")
    else:
        d2 = d2_values[0]
        print(f"\nCalculated diameter d2 = {d2:.4f} mm")

        # Step 4: Socket Outer Diameter (d1)
        d1_squared = (4 * P) / (math.pi * sigma_t_rod) + d2**2
        d1 = math.sqrt(d1_squared)
        print(f"Socket Outer Diameter (d₁): {d1:.4f} mm")

        # Step 5: Collar Diameters
        d3 = 1.5 * d  # Spigot Collar
        d4 = 2.4 * d  # Socket Collar
        print(f"\nSpigot Collar Diameter (d₃): {d3:.4f} mm")
        print(f"Socket Collar Diameter (d₄): {d4:.4f} mm")

        # Step 6: Distances a and c
        a_dist = 0.75 * d
        c_dist = 0.75 * d
        print(f"\nDistance a: {a_dist:.4f} mm")
        print(f"Distance c: {c_dist:.4f} mm")

        # Step 7: Cotter Width (b_cotter)
        b_cotter = P / (2 * t * tau_cotter)
        print(f"Cotter Width (b): {b_cotter:.4f} mm")

        # Step 8: Stress in Spigot End
        sigma_c_spigot = P / (t * d2)
        tau_spigot = P / (2 * a_dist * d2)

        print("\n--- Spigot End ---")
        print(f"Crushing stress σc = {sigma_c_spigot:.2f} N/mm²")
        print(f"Shear stress τ = {tau_spigot:.2f} N/mm²")

        # Step 9: Stress in Socket End
        sigma_c_socket = P / ((d4 - d2) * t)
        tau_socket = P / (2 * (d4 - d2) * c_dist)

        print("\n--- Socket End ---")
        print(f"Crushing stress σc = {sigma_c_socket:.2f} N/mm²")
        print(f"Shear stress τ = {tau_socket:.2f} N/mm²")

        # Step 10: Collar Thickness (t1)
        t1 = 0.45 * d
        print(f"\nCollar Thickness (t₁): {t1:.4f} mm")
else:
    print("Error: No real roots for the quadratic equation.")

```
